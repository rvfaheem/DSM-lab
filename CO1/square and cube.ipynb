```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 10)
y1 =x ** 2
y2 =x ** 3


plt.plot(x, x, 'o-', color='red',label='values')
plt.plot(x, y1, '^--', color='blue',label='square')
plt.plot(x, y2, '*-', color='green',label='cubes')
plt.legend()
plt.grid()
plt.show()
```
