```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
iris=datasets.load_iris()
x=iris.data
y=iris.target
print(x)
print(y)
print(iris.DESCR)
```

```python
print(iris.feature_names)
```

```python
print(iris.target_names)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=2)
print(x_train)
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
y_ans=model.predict(x_test)
print(y_ans)
print(y_test)
```

```python
print(accuracy_score(y_test,y_ans,normalize=False))
```

```python
fig=confusion_matrix(y_test,y_ans,normalize=None,labels=model.classes_)
print(fig)
```

```python
fig_dis=ConfusionMatrixDisplay(confusion_matrix=fig,display_labels=model.classes_)
fig_dis.plot()
plt.show()
```

```python
print(model.predict_proba(x_test))
```
